### Step 2: Data Collection
Use yfinance to download historical data for an NSE stock (e.g., Reliance Industries, which has the ticker symbol 'RELIANCE.NS').


```python
import yfinance as yf
```

```python
# Fetch historical data for a NSE stock
data = yf.download('RELIANCE.NS', start='2010-01-01', end='2023-01-01')
```

### Step 3: Data Preprocessing
Clean and preprocess the data. This includes handling missing values, normalizing data, and creating features.

```python
import pandas as pd
import numpy as np
```

```python
# Fill missing values
data.fillna(method='ffill', inplace=True)
```

```python
# Create new features (e.g., moving averages)
data['MA_50'] = data['Close'].rolling(window=50).mean()
data['MA_200'] = data['Close'].rolling(window=200).mean()
```

```python
# Drop rows with NaN values created by rolling
data.dropna(inplace=True)
```

### Step 4: Create Training and Testing Sets
Split the data into training and testing sets.

```python
# Define the window size for the sequences
window_size = 60

# Prepare the dataset
X = []
y = []

for i in range(window_size, len(scaled_data)):
    X.append(scaled_data.iloc[i-window_size:i].values)
    y.append(scaled_data.iloc[i, 0])

X = np.array(X)
y = np.array(y)

# Split into train and test sets
split = int(len(X) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

### Step 5: Build the Model
Create and compile a neural network model using TensorFlow/Keras.

```python

# Build the LSTM model
model = Sequential()
model.add(Input(shape=(X_train.shape[1], X_train.shape[2])))  # Add Input layer
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=32)

# Predicting the stock prices
predicted_stock_price = model.predict(X_test)

# Inverse transform the predictions to original scale
predicted_stock_price = scaler.inverse_transform(
    np.concatenate((predicted_stock_price, np.zeros((predicted_stock_price.shape[0], 2))), axis=1)
)[:, 0]

# Inverse transform y_test to original scale
real_stock_price = scaler.inverse_transform(
    np.concatenate((y_test.reshape(-1, 1), np.zeros((y_test.shape[0], 2))), axis=1)
)[:, 0]
```

### Step 6: Evaluate the Model
Evaluate the model's performance on the test set.

```python
# Visualize the results
plt.figure(figsize=(14, 7))
plt.plot(data.index[-len(y_test):], real_stock_price, color='red', label='Real Stock Price')
plt.plot(data.index[-len(y_test):], predicted_stock_price, color='blue', label='Predicted Stock Price')
plt.title('Stock Price Prediction')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```
